```python
import folium
m = folium.Map(location=[10.765646150091408, 106.68163817116721], zoom_start=13)

layer1 = folium.FeatureGroup(name='Trường Đại học')
layer2 = folium.FeatureGroup(name='Địa điểm công cộng')


folium.Marker(location=[10.760205, 106.666115],popup="UEH CSB",icon=folium.Icon(color='red',icon='university',prefix='fa')).add_to(layer1)


folium.Marker(location=[10.777000, 106.695300],popup="Dinh Độc Lập",icon=folium.Icon(color='gray',icon='landmark',prefix='fa')).add_to(layer2)
folium.Marker(location=[10.756400, 106.662500],popup="BV ĐHYD",icon=folium.Icon(color='darkred',icon='hospital',prefix='fa')).add_to(layer2)
folium.Marker(location=[10.770833, 106.697778],popup="Metro Bên Thành",icon=folium.Icon(color='yellow',icon='train',prefix='fa')).add_to(layer2)
folium.Marker(location=[10.759975696184588, 106.66727930300873],popup="Sole",icon=folium.Icon(color='brown',icon='coffee',prefix='fa')).add_to(layer2)
folium.Marker(location=[10.774104364369114, 106.69217260979121],popup="công viên Tao Đàn",icon=folium.Icon(color='green',icon='tree',prefix='fa')).add_to(layer2)
layer1.add_to(m)
layer2.add_to(m)

folium.LayerControl().add_to(m)

display(m)
```
