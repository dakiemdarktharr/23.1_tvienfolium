```python
import osmnx as ox
import networkx as nx

g = ox.graph_from_place("District 1, Ho Chi Minh City, Vietnam", network_type='drive')
h = ox.add_edge_speeds(g)
j = ox.add_edge_travel_times(h)

start = ox.nearest_nodes(j, 10.760901065627367, 106.66787922776939)
end = ox.nearest_nodes(j, 10.782929978763931, 106.69487204788973)

path_dijkstra = nx.shortest_path(j, start, end, weight='travel_time', method='dijkstra')
path_astar = nx.astar_path(j, start, end, weight='travel_time')

ox.plot_graph_route(j, path_dijkstra, route_color='red', route_linewidth=5)
ox.plot_graph_route(j, path_astar, route_color='blue', route_linewidth=3)
```
