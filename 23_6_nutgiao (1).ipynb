```python
!pip install osmnx
import osmnx as ox

dia_diem = "District 1, Ho Chi Minh City, Vietnam"
G = ox.graph_from_place(dia_diem, network_type='drive')

thong_ke = ox.basic_stats(G)


chieu_dai_tb = thong_ke['street_length_avg']
print(f"Chieu dai trung binh: {round(chieu_dai_tb, 2)} m")


so_nut = thong_ke['n']
print(f"So luong nut giao: {so_nut}")
```
